```python
import pandas as pd

df_esco = pd.read_csv('data/skills_en_ESCO1.1.csv')#, sep=";")#, on_bad_lines='skip', engine='python')
df_esco
```

```python
escoLabels = df_esco["preferredLabel"].to_list()
escoLabels # 13896 labels!
#len(escoLabels)
```

```python
from sentence_transformers import SentenceTransformer
from sentence_transformers import util
import torch
import pandas as pd

    
def map_Esco(skills, name_model, embeddings):
    model = SentenceTransformer(name_model)
    listLabel1 = []
    listLabel2 = []
    listLabel3 = []
    listLabel4 = []
    listSim1 = []
    listSim2 = []
    listSim3 = []
    listSim4 = []
    df_skills = pd.DataFrame(skills, columns=["originalSkill"])
    for skill in skills:
        #Compute embedding for the skill
        embedding_1= model.encode(skill, convert_to_tensor=True)    # we encode the skill  
        sim = util.pytorch_cos_sim(embedding_1, embeddings)         # get the cosine similarity
        listSim1.append(torch.topk(sim,4)[0][0][0].item())                 # get the 5 maxiumun closest one
        listSim2.append(torch.topk(sim,4)[0][0][1].item())                 # get the 5 maxiumun closest one
        listSim3.append(torch.topk(sim,4)[0][0][2].item())                 # get the 5 maxiumun closest one
        listSim4.append(torch.topk(sim,4)[0][0][3].item())                 # get the 5 maxiumun closest one

        listLabel1.append(escoLabels[torch.topk(sim,4)[1][0][0]])
        listLabel2.append(escoLabels[torch.topk(sim,4)[1][0][1]])
        listLabel3.append(escoLabels[torch.topk(sim,4)[1][0][2]])
        listLabel4.append(escoLabels[torch.topk(sim,4)[1][0][3]])
    print(listLabel1)
    print(listSim1)
    df_skills["esco_label1"] = listLabel1
    df_skills["esco_sim1"] = listSim1
    df_skills["esco_label2"] = listLabel2
    df_skills["esco_sim2"] = listSim2
    df_skills["esco_label3"] = listLabel3
    df_skills["esco_sim3"] = listSim3
    df_skills["esco_label4"] = listLabel4
    df_skills["esco_sim4"] = listSim4
    return df_skills
```

```python
from sentence_transformers import SentenceTransformer
    
# Name of model
name_model= 'sentence-transformers/all-MiniLM-L6-v2'
# We load the model and encode the esco labels (~1 min)
model = SentenceTransformer(name_model)
embeddings = model.encode(escoLabels)
# We call map_Esco function
df_skills = map_Esco(skills, name_model, embeddings)
# We print the results
df_skills
# Copy the results to an excel
df_skills.to_excel("app_all-MiniLM.xlsx")
```

## Atttemp 2: HF -> Sentence Classification

```python
from sentence_transformers import SentenceTransformer, util
import torch
import pandas as pd
from sentence_transformers import SentenceTransformer

# We load the model and encode the esco labels (~1 min)
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
embeddings = model.encode(escoLabels)
print(embeddings)
# We define the function that gets the list of skills and 
# returns the most probable correspondance to ESCO (label, probability and URI)
def mapESCO(skills):
    model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
    listLabel = []
    listSim = []
    listUri = []
    df_skills = pd.DataFrame(skills, columns=["originalSkill"])
    for skill in skills:
        #Compute embedding for the skill
        embedding_1= model.encode(skill, convert_to_tensor=True)    # we encode the skill  
        sim = util.pytorch_cos_sim(embedding_1, embeddings)         # get the cosine similarity
        listSim.append(torch.max(sim).item())                      # get the maximum (the closest one)
        listLabel.append(escoLabels[torch.argmax(sim).item()])      # get the label of the maxiumum
        listUri.append(df_esco["conceptUri"][torch.argmax(sim).item()])  
    df_skills["esco_label"] = listLabel
    df_skills["esco_sim"] = listSim
    df_skills["esco_uri"] = listUri
    return df_skills
        
```

Executing all the skills (~6k), it takes around 6 min

```python
df_skills = mapESCO(skills)
df_skills
```

```python
df_skills.to_excel("app_all-MiniLM.xlsx")
```
